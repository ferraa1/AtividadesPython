```python
#Funções
import math

#Exercício 1
def entre10e80(x):
  if x >= 10 and x <= 80:
    return 'O número digitado está entre 10 e 80'
  else:
    return 'O número digitado não está entre 10 e 80'

#Exercício 2
def categoriaNadador(idade):
  if idade >= 5 and idade <= 7:
    return 'Infantil A'
  elif idade >= 8 and idade <= 10:
    return 'Infantil B'
  elif idade >= 11 and idade <= 13:
    return 'Juvenil A'
  elif idade >= 14 and idade <= 17:
    return 'Juvenil B'
  elif idade >= 18:
    return 'Sênior'
  else:
    return 'Nenhuma'

#Exercício 3
def pesoIdeal(altura,sexo):
  if sexo == 'homem':
    return 72.7 * altura - 58
  else:
    return 62.1 * altura - 44.7

#Exercício 4
#match case requer python 3.10
def mostrarMes(mes):
  if mes == 1:
    return 'Janeiro'
  elif mes == 2:
    return 'Fevereiro'
  elif mes == 3:
    return 'Março'
  elif mes == 4:
    return 'Abril'
  elif mes == 5:
    return 'Maio'
  elif mes == 6:
    return 'Junho'
  elif mes == 7:
    return 'Julho'
  elif mes == 8:
    return 'Agosto'
  elif mes == 9:
    return 'Setembro'
  elif mes == 10:
    return 'Outubro'
  elif mes == 11:
    return 'Novembro'
  elif mes == 12:
    return 'Dezembro'
  else:
    return 'ERROR: Não existe mês com este número!'

#Exercício 5
def raizOuQuadrado(x):
  if x > 0:
    return math.sqrt(x)
  elif x < 0:
    return x * x
  else:
    return 0

#Exercício 6
def calcularMedia(notaUm,notaDois):
  media = (notaUm + notaDois) / 2
  if media >= 7:
    print('Situação: Aprovado. Média:',media)
  elif media < 3:
    print('Situação: Reprovado. Média:',media)
  else:
    print('Situação: Exame. Média:',media)

#Exercício 7
def classeEleitoral(idade):
  if idade < 16:
    return 'Não-eleitor'
  elif idade >= 18 and idade < 65:
    return 'Eleitor obrigatório'
  else:
    return 'Eleitor facultativo'

#Exercício 8
def calcularValor(valor):
  if valor < 30:
    return valor * 1.45
  else:
    return valor * 1.3
```

```python
#Main

#Exercício 1
numero = int(input('Informe um número: '))
print(entre10e80(numero))

#Exercício 2
idade = int(input('Informe a idade do/a nadador/a: '))
print('Categoria:',categoriaNadador(idade))

#Exercício 3
altura = float(input('Informe a altura em metros: '))
sexo = input('Informe o sexo (homem ou mulher): ')
print('Peso ideal:',pesoIdeal(altura,sexo))

#Exercício 4
mes = int(input('Informe um mês em forma de número (1 - 12): '))
print('Mês informado:',mostrarMes(mes))

#Exercício 5
numero = float(input('Informe um número: '))
raizOuQuadrado(numero)

#Exercício 6
notaUm = float(input('Informe a primeira nota: '))
notaDois = float(input('Informe a segunda nota: '))
calcularMedia(notaUm,notaDois)

#Exercício 7
idade = int(input('Informe a idade: '))
print('Classe eleitoral:',classeEleitoral(idade))

#Exercício 8
valor = float(input('Informe o valor do produto: R$'))
print('Valor de venda: R$',calcularValor(valor))
```
