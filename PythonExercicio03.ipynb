```python
#Funções

#1
def segundoValor(lista):
  return lista[1]

#2
def mudarValor(lista):
  final = []
  for x in lista:
    if "maca" in x:
      final.append("laranja")
    else:
      final.append(x)
  return final

#3
def removerValor(lista):
  return [x for x in lista if x != "morango"]

#4
def positivos(lista):
  return [x for x in lista if x > 0]

#5
def multiplicarListas(lista1, lista2):
  result = 1
  for x in (lista1 + lista2):
    result *= x
  return result

def retornoMultiplicacao(resultado):
  return resultado > 40

#6
def mostrarPares(lista):
  return [x for x in lista if x % 2 == 0 and x >= 10 and x <= 50]

#7
def insere(lista, elemento):
  final = lista.copy()
  if elemento in final:
    return final
  else:
    final.append(elemento)
    return final

#8
def retornaSup(lista, numero):
  return [x for x in lista if x > numero]

#9
def fatores(x):
  return [y for y in range(1,(x + 1)) if x % y == 0]

def primo(x):
  return fatores(x) == [1, x]

def primos(lista):
  return [x for x in lista if primo(x)]

#10
def dezA():
  return [x for x in range(0,16) if x % 3 == 0]

def dezB():
  return [x for x in range(0,21) if x % 3 == 0 and x % 2 == 0]

def dezC():
  return [[x] for x in range(1,6)]

def dezD():
  final = []
  for x in range(1,6):
    i = 1
    temp = []
    while i <= x:
      temp.append(1)
      i += 1
    final.append(temp)
  return final

def dezE():
  final = []
  for i in range(1,4):
    for j in reversed(range(1,4)):
      final.append((i,j))
  return final
```

```python
#Main

lista = ["maca", "morango", "uva"]

#1
print(segundoValor(lista))

#2
print(mudarValor(lista))

#3
print(removerValor(lista))

#4
print(positivos([-3, -2, -1, 1, 2, 3]))

#5
resultado = multiplicarListas([1, 2, 3], [2, 2, 2])
if retornoMultiplicacao(resultado):
  print(resultado)
else:
  print("menor que 40")

#6
print(mostrarPares(list(range(1, 50, 3))))

#7
listaParaInserir = list(range(1,20))
print(insere (listaParaInserir, 6))
print(insere (listaParaInserir, 21))
print(insere (listaParaInserir, 69))

#8
print(retornaSup(list(range(1,20)), 13))

#9
print(primos(list(range(1,10))))
print(primos(list(range(1,35))))

#10
print(dezA())
print(dezB())
print(dezC())
print(dezD())
print(dezE())
```
