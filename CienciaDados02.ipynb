```python
#1
import matplotlib.pyplot as plt
import pandas
import seaborn

df = pandas.read_csv('MATRICULAS.csv')

ano = df['ANO']
matriculas = df['MATRICULAS']
evasao = df['EVASAO']

df['PERCENTUAL DE EVASAO'] = evasao * 100 / matriculas

print(df)

plt.title('Distribuição de Matrículas e Evasão')
plt.xlabel('Ano')
plt.ylabel('Matrículas e Evasão')
plt.bar(ano, matriculas)
plt.bar(ano, evasao)
plt.show()
plt.close()

grafico = seaborn.scatterplot(data=df,x=df['ANO'],y=df['PERCENTUAL DE EVASAO'])
plt.title('Distribuição de Percentual de Evasão')
plt.show()
plt.close()
```

```python
#2
import matplotlib.pyplot as plt
import numpy
import pandas
import seaborn
import statistics

df = pandas.read_csv('TEMPERATURAS.csv')

minimas = df['MINIMA']
maximas = df['MAXIMA']

print(df)

plt.boxplot(minimas)
plt.title("Mínimas - BoxPlot")
plt.show()
plt.close()

q1 = numpy.percentile(minimas, 25)
q3 = numpy.percentile(minimas, 75)
print("1o. Quartil --->", q1)
print("Mediana ------->", statistics.median(minimas))
print("3o. Quartil --->", q3)
print("Mínimo -------->", q1 - 1.5 * (q3 - q1))
print("Máximo -------->", q3 + 1.5 * (q3 - q1))

plt.boxplot(maximas)
plt.title("Máximas - BoxPlot")
plt.show()
plt.close()

q1 = numpy.percentile(maximas, 25)
q3 = numpy.percentile(maximas, 75)
print("1o. Quartil --->", q1)
print("Mediana ------->", statistics.median(maximas))
print("3o. Quartil --->", q3)
print("Mínimo -------->", q1 - 1.5 * (q3 - q1))
print("Máximo -------->", q3 + 1.5 * (q3 - q1))
```

```python
#5
import csv
import numpy
import pandas

with open('anotacoes.txt', 'r') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split("\t") for line in stripped if line)
    with open('anotacoes.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerow(('col1','col2','col3','col4','col5','col6','col7'))
        writer.writerows(lines)

df = pandas.read_csv('anotacoes.csv')

print(df)

lista = []
lista = list(df['col1']) + list(df['col2']) + list(df['col3']) + list(df['col4']) + list(df['col5']) + list(df['col6']) + list(df['col7'])

mediana = numpy.median(lista)
print("Mediana:",mediana)
medianas = []
medianas.append(numpy.median(df['col1']))
medianas.append(numpy.median(df['col2']))
medianas.append(numpy.median(df['col3']))
medianas.append(numpy.median(df['col4']))
medianas.append(numpy.median(df['col5']))
medianas.append(numpy.median(df['col6']))
medianas.append(numpy.median(df['col7']))
mediaMediana = numpy.mean(medianas)
print("Média Aritmética da Mediana de 7 grupos:",mediaMediana)

plt.boxplot(lista)
plt.title("Circunferências - BoxPlot")
plt.show()
plt.close()
q1 = numpy.percentile(lista,25)
print("1o. Quartil -->",q1)
print("Mediana ------>",mediana)
q3 = numpy.percentile(lista,75)
print("3o. Quartil -->",q3)
minimo = q1 - 1.5 * (q3 - q1)
print("Mínimo ------->",minimo)
maximo = q3 + 1.5 * (q3 - q1)
print("Máximo ------->",maximo)

tratado = [x for x in lista if x >= minimo and x <= maximo]
print("Quantidade de elementos original:",len(lista))
print("Quantidade de elementos tradado:",len(tratado))

colunas = []
for x in range(32):
  colunas.append(list(df.iloc[x]))

plt.title("Circunferências - Dispersão")
grafico = seaborn.scatterplot(data=colunas, legend=None)
plt.show()
plt.close()

primeiro = []
segundo = []
meio = []
terceiro = []
quarto = []
for x in tratado:
  if (x == mediana):
    meio.append("Mediana")
  elif (x >= minimo and x < q1):
    primeiro.append("1o. Quarto")
  elif (x >= q1 and x < mediana):
    segundo.append("2o. Quarto")
  elif (x > mediana and x <= q3):
    terceiro.append("3o. Quarto")
  elif (x > q3 and x <= maximo):
    quarto.append("4o. Quarto")
quantidade = primeiro + segundo + meio + terceiro + quarto

plt.title("Circunferências - Histograma")
plt.xlabel('Localização')
plt.ylabel('Quantidade')
plt.hist(quantidade, bins=5, align='mid')
plt.show()
plt.close()
```
