```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df1 = pd.read_csv('tennis_stats.csv')

df1X = np.array(df1['Wins']).reshape(-1,1)
df1Y = df1['Winnings']

df1RegressaoLinear = LinearRegression().fit(df1X, df1Y)
df1Previsao = df1RegressaoLinear.predict(df1X)
plt.plot(df1X, df1Y, 'o')
plt.plot(df1X, df1Previsao)
plt.show()

df1R2 = r2_score(df1Y, df1Previsao)
print("R^2=",df1R2)

###

df2 = pd.read_csv('elevation_suicide_2018.csv')

df2X = np.array(df2['Mean Elevation (meters)'])
df2Y = np.array(df2['Suicide Death Rate (%)'])

#Exponencial:
#df2Regressao = np.polyfit(df2X, np.log(df2Y), 1)
#Logaritmica:
df2Regressao = np.polyfit(np.log(df2X), df2Y, 1)
df2Previsao = df2Regressao.predict(df1X)
plt.plot(df2X, df2Y, 'o')
plt.plot(df2X, df2Previsao)
plt.show()

print(df2Regressao)
df2R2 = r2_score(df2Y, df2Previsao)
print("R^2=",df2R2)
```
